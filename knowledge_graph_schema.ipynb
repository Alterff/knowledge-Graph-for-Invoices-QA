```python
def extract_industry(json_str):
    try:
        data = json.loads(json_str)
        return data.get('industry', None)
    except json.JSONDecodeError:
        return None

def extract_languages(json_list):
    try:
        languages = [entry['title'] for entry in json.loads(json_list)]
        return '|'.join(languages)
    except:
        return None

def extract_country(string):
    if isinstance(string, str):
        elements = string.split(',')
        return elements[-1].strip()
    else:
        return None

df['industry'] = df['current_company'].apply(lambda x: extract_industry(x))
df['languages'] = df['languages'].apply(lambda x: extract_languages(x))
df['country'] = df['city'].apply(lambda x: extract_country(x))
df = df [['id','name','current_company:name','educations_details','languages','industry','country']].dropna()
industry_counts = df['industry'].value_counts()
df = df[df['industry'].isin(industry_counts[industry_counts > 2].index)].reset_index(drop=True)
df = df.rename(columns={'current_company:name': 'company','educations_details':'education'})
df.head(10)
```

```python
import pandas as pd
from neo4j import GraphDatabase




# Define the function to load data into Neo4j
df['product_description'] = df['product_description'].astype(str)
df['quantity'] = df['quantity'].astype(str)
df['unit_price'] = df['unit_price'].astype(str)
df['total'] = df['total'].astype(str)
def load_data(tx, row):
    # Extract static data
    invoice_id = row['InvoiceID']
    client = row['client']
    seller = row['seller']
    total_amount = row['Total_amount']

    # Split the pipe-separated values
    product_descriptions = row['product_description'].split('|')
    quantities = row['quantity'].split('|')
    unit_prices = row['unit_price'].split('|')
    totals = row['total'].split('|')

    # Construct the Cypher query dynamically
    query = f"""
    // Create Invoice Node
    MERGE (invoice:Invoice {{name: $invoice_id}})

    // Create Client Node
    MERGE (client:Client {{name: $client}})

    // Create Seller Node
    MERGE (seller:Seller {{name: $seller}})

    // Create Relationships
    MERGE (invoice)-[:HAS_CLIENT]->(client)
    MERGE (invoice)-[:HAS_SELLER]->(seller)

    // Associate the total amount with the invoice
    MERGE (invoice)-[:HAS_TOTAL_AMOUNT]->(:TotalAmount {{name: $total_amount}})
    """

    # Dynamic Cypher for product-related data
    for i, prod in enumerate(zip(product_descriptions, quantities, unit_prices, totals)):
        product_desc, quantity, unit_price, total = prod


        query += f"""
        // Create and associate product details for product {i+1}
        MERGE (product_desc_{i}:ProductDescription {{name: $product_desc_{i}}})
        MERGE (invoice)-[:HAS_PRODUCT]->(product_desc_{i})
        MERGE (product_desc_{i})-[:HAS_QUANTITY]->(:Quantity {{name: $quantity_{i}}})
        MERGE (product_desc_{i})-[:HAS_UNIT_PRICE]->(:UnitPrice {{name: $unit_price_{i}}})
        MERGE (product_desc_{i})-[:HAS_TOTAL]->(:Total {{name: $total_{i}}})
        """

    # Parameters for the Cypher query
    parameters = {
        'invoice_id': invoice_id,
        'client': client,
        'seller': seller,
        'total_amount': total_amount
    }

    # Add dynamic parameters for each product
    for i, prod in enumerate(zip(product_descriptions, quantities, unit_prices, totals)):
        parameters[f'product_desc_{i}'] = prod[0]
        parameters[f'quantity_{i}'] = prod[1]
        parameters[f'unit_price_{i}'] = prod[2]
        parameters[f'total_{i}'] = prod[3]

    # Run the query with the parameters
    tx.run(query, parameters)



# Connect to Neo4j and load data
driver = GraphDatabase.driver(neo4j_url, auth=(neo4j_user, neo4j_password))

with driver.session() as session:
    for index, row in df.iterrows():
        session.write_transaction(load_data, row)

driver.close()
```

```python
from neo4j import GraphDatabase
def load_data(tx, row):
    query = """
    MERGE (person:Person {name: $name})
    MERGE (company:Company {name: $company})
    MERGE (school:School {name: $education})
    MERGE (industry:Industry {name: $industry})
    MERGE (country:Country {name: $country})
    FOREACH (lang IN split($languages, '|') |
        MERGE (language:Language {name: trim(lang)})
        MERGE (person)-[:SPEAKS]->(language))
    MERGE (person)-[:WORKS_IN]->(company)
    MERGE (person)-[:LIVES_IN]->(country)
    MERGE (person)-[:EDUCATED_AT]->(school)
    MERGE (company)-[:IS_IN]->(industry);
    """
    tx.run(query,
           name=row['name'],
           company=row['company'],
           education=row['education'],
           industry=row['industry'],
           country=row['country'],
           languages=row['languages'])

# Connect to Neo4j and load data
driver = GraphDatabase.driver(neo4j_url, auth=(neo4j_user, neo4j_password))

with driver.session() as session:
    for index, row in df.iterrows():
        session.write_transaction(load_data, row)

driver.close()
```
